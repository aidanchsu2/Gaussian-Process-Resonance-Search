```python
import matplotlib.pyplot as plt 
import pandas as pd
import math
file_1 = pd.read_excel("1_sigma_rbn40_no_signal.xlsx")
x1 = file_1["Mass"]
y1 = file_1["Chisquare Probability"]
plt.plot(x1,y1)
plt.ylabel("p-value")
plt.xlabel("Mass")
plt.title("1 sigma, rebin 40, no signal")
```

```python
file_2 = pd.read_excel("1_sigma_rbn10_no_signal.xlsx")
x2 = file_2["Mass"]
y2 = file_2["Chisquare Probability"]

plt.plot(x1,y1,x2,y2)
plt.ylabel("p-value")
plt.xlabel("Mass")
plt.title("1 sigma, rebin 10, no signal")
```

```python
logy2 = []
for i in y2:
    if i == 0:
        logy2.append(0)
    else:
        logy2.append(math.log(i,10))
plt.plot(x2,logy2)
```
